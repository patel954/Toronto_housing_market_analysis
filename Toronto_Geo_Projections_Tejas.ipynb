# Toronto Area Geographic Projections

```python
# Import the required libraries and dependencies
import pandas as pd
import os
import hvplot.pandas
from pathlib import Path
```

```python
# Using the read_csv function and Path module, create a DataFrame 
sfo_data_df = pd.read_csv(Path('../Toronto_housing_market_analysis/Data/C01_C15_Clean_df.csv'))
sfo_data_df.set_index('Location', inplace=True)

# Review the first and last five rows of the DataFrame
display(sfo_data_df.head())
display(sfo_data_df.tail())
```

```python
#Spliting year, month and date from date column
realestate_to_df = sfo_data_df
realestate_to_df[["year", "month", "day"]] = sfo_data_df['Date'].str.split("-", expand = True)
realestate_to_df
```

```python
# Create a numerical aggregation that groups the data by the year and then averages the results
all_neighbour = sfo_data_df.groupby("Location").mean()
all_neighbour = all_neighbour[["CompBenchmark", "SFDetachBenchmark", "ApartBenchmark"]]

# Review the DataFrame
all_neighbour
```

```python
# Load neighborhoods coordinates data
neighborhood_locations_df = pd.read_csv(Path("/Users/tejaspatel/Documents/Toronto_housing_market_analysis/Data/Toronto_MLS_Lat_Lon.csv", index_col="Location"))
neighborhood_locations_df.set_index('Location', inplace=True)

# Review the DataFrame
neighborhood_locations_df
```

```python
#Combining both the dataframe 
all_neighborhoods_df = pd.concat(
    [neighborhood_locations_df, all_neighbour],
    axis="columns",
    sort=False 
)
#Display the combined datafram
all_neighborhoods_df
```

```python
# Create a plot to analyze neighborhood info of Apartments
all_neighborhoods_df.hvplot.points(
    'Lon',
    'Lat',
    geo=True,
    size=3000,
    color='ApartBenchmark',
    tiles='OSM',
    frame_width=700,
    frame_height=500,
    title=''
)
```

```python
# Create a plot to analyze neighborhood info on Semi-detach houses
all_neighborhoods_df.hvplot.points(
    'Lon',
    'Lat',
    geo=True,
    size=3000,
    color='SFDetachBenchmark',
    tiles='OSM',
    frame_width=700,
    frame_height=500,
    title=''
)
```

```python
# Create a plot to analyze neighborhood info of Composite Benchmark
all_neighborhoods_df.hvplot.points(
    'Lon',
    'Lat',
    geo=True,
    size=3000,
    color='CompBenchmark',
    tiles='OSM',
    frame_width=700,
    frame_height=500,
    title=''
)
```
